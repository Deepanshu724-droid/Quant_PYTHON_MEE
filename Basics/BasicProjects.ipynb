## DICE SIMULATIONS

```python
print("Welcome to dice simulation")
DICE_FACE = [1,2,3,4,5,6]
DICE_no=float(input("how many times you want to roll a dice ? "))
import random 
i= 1 
while i<= DICE_no:
    res = random.choice(DICE_FACE)
    print(res)
    i+=1 
    
    
```

# Stock Market %change Calculator

```python
Buy_price = float(input("Enter buying price"))
Selling_price = float(input("Enter Selling price"))
Cal = ((Selling_price - Buy_price )/ Buy_price) * 100 

if Cal > 0 :
    print("profit")
elif Cal < 0 :
    print("loss")
else:
    print("No lose No profit, net zero")
print(f"Percentage Change : {Cal:.2f}%")
```
